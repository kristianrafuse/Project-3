```python
import geopy.geocoders
from geopy.exc import GeocoderTimedOut

# Set custom user agent
custom_user_agent = "my-forest-app"

# Create a geocoder object with the custom user agent
geolocator = geopy.geocoders.Nominatim(user_agent=custom_user_agent)

# Define a function to geocode country names and get coordinates
def geocode_country(country):
    try:
        location = geolocator.geocode(country)
        if location is not None:
            return location.latitude, location.longitude
        else:
            return None, None
    except GeocoderTimedOut:
        return geocode_country(country)  # Retry if geocoding times out

# Apply geocoding function to dataframe and get latitude and longitude
forests_df_percent['Latitude'], forests_df_percent['Longitude'] = zip(*forests_df_percent['Country'].apply(geocode_country))
```
